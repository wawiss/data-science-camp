## Control flow

*   if
*   for
*   while



```python
#if
score =82
if score>=80 :
    print("pass")
else:
     print("fail")
```

```python
#multiple grade
if score >= 80:
    grade="A"
elif score>=70:
    grade = "B"
elif score>=60:
    grade="C"
else:
    grade= "D"

print(grade)
```

```python
#get input from user
#input recieve in string format only!!
username= input("what's your name :")
age=int(input("How olds are you : "))
```

```python
#multiple condition
#night club
has_id=True
age_over_18 = False
if has_id and age_over_18:
    print("welcome to the club")
else:
    print("sorry you cant come in.")
```

```python
vip_customer=True

if has_id and (age_over_18 or vip_customer) :
    print("welcome to the club")
else:
    print("sorry you cant come in.")
```

For & while

```python
#for
items=["egg","milk","cake"]

for item in items:
    if item=="milk":
        print("CP meiji")
    else:
        print(item)
```

```python
#want to uppercase the list
result=[]
for item in items:
    result.append(item.upper())
print(result)
```

```python
#list comprehention
[item.upper() for item in items]
```

```python
#while loop
while(1):
    user_input=input("Play on : Y/N")
    if user_input=="Y":
        print("lets go")
    else:
        print("Good bye")
        break
```

```python
count=0
while( count<=5):
    print("hi")
    count+=1
```
