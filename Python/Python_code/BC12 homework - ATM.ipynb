```python
from IPython.core.completer import ProvisionalCompleterWarning
class ATM():
    def __init__(self,username,bank,amount):
        self.username = username
        self.bank = bank
        self.amount=amount

    #string methods -> show data in object
    def __str__(self) :
        return f"{self.username} using {self.bank} and has balance {self.amount} in account"

    #check balance function
    def check_balance(self):
        print(f"balance : {self.amount} bath")

    #deposit function
    def deposit(self,saving):
        self.amount += saving
        print(f"Deposit {saving} THB to {self.username} sucessfully")

    #withdraw function
    def withdraw(self,withdraw_amount):
        if self.amount < withdraw_amount:
            print("Remaining balance not enough for withdraw")
        else :
            self.amount -= withdraw_amount
            print(f"sucessfully withdraw {withdraw_amount} from {self.username} account. \n Remaining balance : {self.amount}")

    #transfer money
    def transfer(self,destination_account,transfer_amount):
        if self.amount < transfer_amount:
            print("Transfer failed.Remaining balance not enough.")
        else:
            destination_account.deposit(transfer_amount)
            self.amount -= transfer_amount
            print(f"Sucessfully transfer {transfer_amount} from {self.username} to {destination_account.username}")
            print(f"Remaining balance: {self.amount} THB")

    #pay bill
    def paybill(self,bill_no,bill_amount):
         if self.amount < bill_amount:
            print("Pay bill failed. Remaining balance not enough.")
         else:
            self.amount -= bill_amount
            print(f"Sucessfully paid bill : {bill_no}")
            print(f"Remaining balance: {self.amount} THB")

    #Cash Exchange
    def cash_exchange(self,exchange_amount,currency_type):
        if self.amount < exchange_amount:
            print("Exchange failed.Remaining balance not enough.")
        else:
            if currency_type =="USD":
                currency_exchange =0.030 * exchange_amount
            elif currency_type =="JPY" :
                currency_exchange = 4.84 * exchange_amount
            elif currency_type=="CNY":
                currency_exchange = 0.21 * exchange_amount
            else :

                return("Invalid currency type.")

            self.amount -=exchange_amount
            print(f"Sucessfully exchange to {currency_exchange} {currency_type}")
            print(f"Remaning balance : {self.amount}")

```
