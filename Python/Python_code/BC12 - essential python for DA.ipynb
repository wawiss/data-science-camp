## OOP

teach python to understand object
*   attribute : attribute of object
*   methods : function for object



```python
class Dog:
    #attribute
    def __init__(self,name,species,age):
        self.name=name
        self.species=species
        self.age=age

    #function
    def bark(self):
        print(f"{self.name} is hong hong")

    def hbd(self):
        self.age+=1
        print(f"{self.name} is now {self.age} years old")
```

```python
dog1=Dog("milo","siberian husky",2)
dog2=Dog("david","Shiba",3)
```

```python
print(f"{dog1.name} is {dog1.species}")
```

```python
dog2.bark()
```

```python
dog1.hbd()
```

OOP-Super class

class that extend from other class

```python
class Person:
    def __init__ (self,name,age):
        self.name=name
        self.age=age

#extension
class Employee(Person):
    def __init__(self, name, age,company):
        #refer to Super class
        super().__init__(name, age)
        self.company=company

    def greeting(self):
        print("hello world")
```

```python
person1=Person("bright",23)
```

```python
#person cant used employee func
person1.greeting()
```

```python
person2=Employee("Lisa",37,"Black pink")
```

```python
person2.greeting()
```

```python
#read file
import csv
```

```python
#file not found error
try:
    file=open("hotel.csv","r") #r w a
    #csv.reader(csv_file)
    data =csv.reader(file)
    for row in data:
        print(row)
    #if not close -> oppunity to corrupt
    file.close()
except FileNotFoundError:
    print("file not found, please check filename again.")
# at the end
finally:
    print("read file succesfully")
```

```python
#context manager
with open("hotel.csv","r") as file:
    data= csv.reader(file)
    for row in data:
        print(row)
```

```python
#context manager ver 2
read_data=[]
with open("hotel.csv","r") as file:
    data= csv.reader(file)
    for row in data:
        read_data.append(row)
```

```python
#header
read_data[0]
```

```python
data_count=len(read_data)
```

```python
#for loop to calculate avg,price per night
price=[]
for row in read_data[1:]:
    price.append(int(row[-1]))

avg_price_per_night=sum(price)/data_count
print(f"average price per night is {avg_price_per_night}")
```

```python
price=0
for row in read_data[1:]:
    price += (int(row[-1]))

avg_price_per_night=price/data_count
print(f"average price per night is {avg_price_per_night}")
```

```python
##for loop : get data from api for person 1 to 10
import time

name=[]
for i in range(10):
    url=f"https://swapi.info/api/people/{i+1}"
    response=requests.get(url)
    response_json=response.json()
    name.append(response_json["name"])
    print(f"sucess character :{i+1}")
    time.sleep(1)

print(name)
```

```python
##for loop : get data from api for person 1 to 10
import time

characters=[]
for i in range(10):
    url=f"https://swapi.info/api/people/{i+1}"
    response=requests.get(url)
    response_json=response.json()
    data=[response_json["name"],
          response_json["height"],
          response_json["mass"]]
    characters.append(data)
    print(f"sucess character :{i+1}")
    time.sleep(1)

print(characters)
```
