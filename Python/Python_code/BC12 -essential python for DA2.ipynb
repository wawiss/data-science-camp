## Map/filter

map

```python
scores=[82,75,80,66,50,49]
```

```python
#map/filter
#iterable

for score in scores:
    if score >= 80:
        print("pass")
    else:
        print("fail")
```

```python
#map function
def grading(scores):
    if scores>= 80:
        print("pass")
    else:
        print("fail")
```

```python
#map
list(map(grading,scores))
```

```python
#add five to scores
list(map(lambda x : x+5,scores))
```

filter

```python
##filter
scores
```

```python
list(filter(lambda x : x>=80 ,scores))
```

```python
friend_scores=[["wawiss",80],["kuroo",98],["bokuto",57]]
#filter score >=80
list(filter(lambda x: x[1]>=80,friend_scores))
```
