# Harry Potter API

```python
#import library
import requests
import pandas as pd
```

## Harry Potter Books

```python
#Book api
url="https://potterapi-fedeperin.vercel.app/en/books"
response=requests.get(url)
book=response.json()
book
```

```python
#create book dataframe
book_df=pd.DataFrame(book,columns=["number","title","originalTitle","releaseDate","description","pages","cover"])
book_df
```

## Harry Potter Characters

```python
url="https://potterapi-fedeperin.vercel.app/en/characters"
response=requests.get(url)
characters=response.json()
characters
```

```python
#Create character dataframe
character_df=pd.DataFrame(characters,columns=["index","fullName","nickname"
,"hogwartsHouse","interpretedBy","children"
,"image","birthdate"])
character_df
```
